### Find out the IQR for the column - Age 

*Hint: IQR - Interquartile range = 75th percentile - 25th percentile* 

```python
IQR = HRDF['Age'].quantile(0.75)-HRDF['Age'].quantile(0.25)
print (IQR)
```

### Find out the mean, median and variance for the column - DailyRate 

```python
DailyRate_mean = HRDF['DailyRate'].mean()
DailyRate_median = HRDF['DailyRate'].median()
DailyRate_var = HRDF['DailyRate'].var()
print("DailyRate Mean: "+ str(DailyRate_mean)+" DailyRate Median "+ str(DailyRate_median)+" DailyRate Variance "+str(DailyRate_var) )
```
