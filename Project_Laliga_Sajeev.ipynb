```python
import numpy as np
import pandas as pd
```

```python
La_liga=pd.read_csv('Laliga.csv')
Laliga=pd.DataFrame(La_liga)
```

Reading the data-set and replacing dashes with zero so as to make the set ready for arithmetic operation

```python
Laliga=Laliga.replace('-',0)
```

Teams which started playing between 1930 and 1980

```python
Laliga['DebutYear']=Laliga['Debut'].astype(str).str[:4]
Laliga['DebutYear']=Laliga['DebutYear'].astype(int)
Debut3080 = Laliga[(Laliga['DebutYear']>1929)&(Laliga['DebutYear']<1981)]
print(Debut3080[['Team','DebutYear']])
```

List of teams which came top 5 in terms of points

```python
Laliga[["Pos","Points"]]=Laliga[["Pos","Points"]].apply(pd.to_numeric)
Laliga.sort_values('Points', ascending = False)
print(Laliga.loc[0:4,'Team'])
```

Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. Using the same function, find the team which has maximum and minimum goal difference.

```python
Laliga[["GoalsFor","GoalsAgainst"]]=Laliga[["GoalsFor","GoalsAgainst"]].apply(pd.to_numeric)
def Goal_diff_count():
    Laliga['GoalDiff']=Laliga['GoalsFor']-Laliga['GoalsAgainst']
    GDTable=pd.DataFrame()
    GDTable=Laliga[['Team','GoalDiff']]
    print(GDTable)
    MaxGD=GDTable['GoalDiff'].idxmax()
    print(GDTable.loc[MaxGD])
    MinGD=GDTable['GoalDiff'].idxmin()
    print(GDTable.loc[MinGD])
    return
Goal_diff_count()
```

Create a new column with name “Winning Percent” and append it to the data set 

```python
Laliga[["GamesPlayed","GamesWon"]]=Laliga[["GamesPlayed","GamesWon"]].apply(pd.to_numeric)
Laliga['Winning_Percent']=(Laliga['GamesWon']/Laliga['GamesPlayed'])*100
# Complete Table
print(Laliga[['Team','GamesPlayed','Winning_Percent']])
# Top 5 teams by Winning Percent
Laliga.sort_values('Winning_Percent', ascending = False)
print(Laliga.loc[0:4,'Team'])
```

Group teams based on their “Best position” and print the sum of their points for all positions

```python
Laliga.groupby('BestPosition')['Points'].sum()
```
