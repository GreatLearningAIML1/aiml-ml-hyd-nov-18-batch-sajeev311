```python
import numpy as np
from scipy.stats import ttest_1samp, ttest_ind, mannwhitneyu, levene, shapiro
from statsmodels.stats.power import ttest_power
import pandas as pd
```

```python
titan=pd.read_csv("Titan_Insurance.csv")
```

Converting the Data from Thousands, and dropping the earlier columns




```python
titan['Old']=titan['Old Scheme (in thousands)']*1000
titan['New']=titan['New Scheme (in thousands)']*1000
```

```python
titan = titan.drop('Old Scheme (in thousands)',1)
titan = titan.drop('New Scheme (in thousands)',1)
```

# Q1. Find the mean of old scheme and new scheme column.

```python
# Mean of old scheme
print("Mean of the Old Scheme was", titan['Old'].mean())
# Mean of new scheme
print("Mean of the New Scheme is", titan['New'].mean())
```

# Q2. Use the five percent significance test over the data to determine the p value to check new scheme has significantly raised outputs? 

Null Hypothesis: Mean of Old Scheme = Mean of New Scheme
Alternate Hypothesis: Mean of Old Scheme < Mean of New Scheme

We need to apply 2-sample paired t-test (which is the same test as a one sample t-test in Scipy) to determine if new scheme has significantly increase output

```python
t_statistic, p_value = ttest_1samp(titan['New']-titan['Old'], 0)
print(t_statistic, p_value)
```

# Q3. What conclusion does the test (p-value) lead to?

The p-value of 0.13 is greater than 0.05, which means that we fail to reject the null hypothesis. This means that the New Scheme hasn't significantly raised the outputs

# Probability of Type-1 error

Since Confidence interval 95%, alpha is .05, so probability of Type-I error is 5%

# What is the p- value of the hypothesis test if we test for a difference of 5000

```python
t_statistic, p_value = ttest_1samp(titan['New']-titan['Old'], 4999)
print(t_statistic, p_value)
```

Since p_value is greater than 0.05, we fail to reject the null hypothesis that the difference of means is less than 5000
